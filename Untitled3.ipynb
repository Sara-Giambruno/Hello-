```python
from plotnine import *
from gapminder import gapminder
```

```python
(
ggplot(gapminder)
+ aes(x="continent", y="lifeExp", color="continent")
+ geom_boxplot()
+ facet_wrap("~year")
+ labs(x="Continent", y="Life Expectancy", title="Life Expectancy by Continent")
+ theme_minimal()
)
```

```python
(
ggplot(gapminder)
+ geom_line(aes(x="year", y="lifeExp", group="country"), alpha=0.3)
+ geom_smooth(aes(x="year", y="lifeExp", color="continent"), se=False)
+ facet_wrap("~continent")
+ theme_minimal()
)
```
